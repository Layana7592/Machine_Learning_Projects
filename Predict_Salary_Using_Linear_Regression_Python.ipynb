Importing Libraries and Reading the Salary Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("/content/Salary_Data.csv")
df
```

```python
df.head()
```

Initial Data Exploration: Head, Tail, and Dataset Summary

```python
df.tail()
```

```python
df.info()
```

```python
df.dtypes
```

Null Value Analysis Across the Dataset

```python
df.isna().sum()
```

Extracting Independent and Dependent Variables

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
y
```

Data Visualization: Experience vs Salary Scatter Plot

```python
plt.scatter(x,y,color='r')
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Experience vs Salary")
```

Preparing Training and Testing Data Using train_test_split()

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

Fitting the Linear Regression Model and Generating Predictions

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

Evaluation Table: Actual Values, Predicted Values, and Their Differences

```python
df1=pd.DataFrame({'Actual_value ':y_test,
                  'Predicted_value ':y_pred,
                  'Difference':y_test-y_pred})
df1
```

Model Inference: Predicting Output for New Input Value

```python
print(model.predict([[5.8]]))
```

Regression Visualization: Model Fit on Training and Test Data

```python
plt.scatter(x_train,y_train)
plt.plot(x_test,y_pred,color='r')
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Experience vs Salary")
```

Displaying Linear Regression Coefficient and Constant

```python
print("Slop is :",model.coef_)
```

```python
print("Constant is",model.intercept_)
```

Performance Metrics: Mean Absolute Error, Mean Absolute Percentage Error, and R Score

```python
from sklearn.metrics import mean_absolute_error
print("Error is :",mean_absolute_error(y_test,y_pred))
```

```python
from sklearn.metrics import mean_absolute_percentage_error
print("MAPE is:",mean_absolute_percentage_error(y_test,y_pred))
```

```python
from sklearn.metrics import r2_score
print("r2 score is :",r2_score(y_test,y_pred))
```
